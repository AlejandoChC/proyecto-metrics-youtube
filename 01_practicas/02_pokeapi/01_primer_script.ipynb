```python
# Tipos (type) del Pokémon (puede tener uno o dos tipos)

pokemon_especifico = requests.get("https://pokeapi.co/api/v2/pokemon/bulbasaur")

datos = pokemon_especifico.json()

tipos = datos["types"]

print(len(tipos))

for tipo_en_info in tipos:
    nombre_tipo = tipo_en_info["type"]["name"]
    nombre_slot = tipo_en_info["slot"]
    print(f"{nombre_slot} - {nombre_tipo}")
```

```python
# Habilidades (abilities):

habilidad = datos["abilities"]
print(len(habilidad))
print(habilidad)

lista_habilidades =[]
for habilidad_interna in habilidad:
    nombre_habilidad = habilidad_interna["ability"]["name"]
    lista_habilidades.append(nombre_habilidad)
print(lista_habilidades)
```

```python
# Estadísticas base (stats): 

estadisticas = datos["stats"]
print(len(estadisticas))
print(estadisticas)

for estat_interna in estadisticas:
    nombre_base = estat_interna["base_stat"]
    nombre_stat = estat_interna["stat"]["name"]
    print(f"estadistica: {nombre_stat} - valor base: {nombre_base}")
```
